# Personal Expense Tracker

Import necessary modules

```python
import csv
```

Create a list and global variable to store data

```python
expenses = [] # To store all expenses
budget = 0.0 # To store the monthly budget
```

Function to add an expense


```python
def add_expense():
    date = input("Enter the date of the expense (YYYY-MM-DD): ")
    category = input("Enter the category of the expense (e.g., Food, Travel): ")
    amount = float(input("Enter the amount of the expense: "))
    description = input("Enter a brief description of the expense: ")

    # Store the expense as a dictionary
    expense = {'date': date, 'category': category, 'amount': amount, 'description': description}
    expenses.append(expense)

    print("\nExpense successfully added!\n")
```

Function to view all expenses

```python
def view_expenses():
    if not expenses:
        print("No expenses found!\n")
        return

    print("\nAll Expenses:")
    for expense in expenses:
        print(f"Date: {expense['date']}, Category: {expense['category']}, Amount: {expense['amount']}, Description: {expense['description']}")
    print("\n")
```

Function to set and track the monthly

```python
# Function to set the monthly budget
def set_budget():
    global budget
    budget = float(input("Enter your monthly budget: "))
    print(f"\nYour monthly budget of {budget} has been set!\n")

# Function to track the budget
def track_budget():
    if budget == 0.0:
        print("Please set the monthly budget first.\n")
        return

    total_expense = sum(expense['amount'] for expense in expenses)

    if total_expense > budget:
        print(f"\nYou have exceeded your budget by {total_expense - budget}!\n")
    else:
        print(f"\nYou have {budget - total_expense} left for the month.\n")
```

Function to save and load expenses from a CSV file

```python
# Function to save expenses to a CSV file
def save_expenses_to_file():
    with open('expenses.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Date', 'Category', 'Amount', 'Description']) # Header row
        for expense in expenses:
            writer.writerow([expense['date'], expense['category'], expense['amount'], expense['description']])
    print("\nAll expenses successfully saved to the file!\n")
```

```python
# Function to load expenses from a CSV file
def load_expenses_from_file():
    try:
        with open('expenses.csv', mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                expenses.append({
                    'date': row['Date'],
                    'category': row['Category'],
                    'amount': float(row['Amount']),
                    'description': row['Description']
                })
        print("\nExpenses successfully loaded from the file!\n")
    except FileNotFoundError:
        print("\nNo expense file found, starting a new record.\n")
```

Function for the interactive menu

```python
# Menu function
def menu():
    while True:
        print("Expense Tracker Menu:")
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Set Budget")
        print("4. Track Budget")
        print("5. Save Expenses")
        print("6. Load Expenses")
        print("7. Exit")
        choice = input("Choose an option: ")

        if choice == '1':
            add_expense()
        elif choice == '2':
            view_expenses()
        elif choice == '3':
            set_budget()
        elif choice == '4':
            track_budget()
        elif choice == '5':
            save_expenses_to_file()
        elif choice == '6':
            load_expenses_from_file()
        elif choice == '7':
            print("Exiting...")
            break
        else:
            print("Invalid option. Please try again.\n")

# Load expenses first and then display the menu
load_expenses_from_file()
menu()
```
